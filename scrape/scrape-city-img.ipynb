```python
import wikipedia
import urllib
import pandas as pd
```

```python
PROJECT_DIR = "D:/analytics/shiny/nomad_life/"
data = pd.read_csv(PROJECT_DIR + "/data/city_list.txt")
data.head()
```

```python
PAGE = ['Warsaw']

wikipage = wikipedia.page(PAGE)
print("Page Title: %s" % wikipage.title)
print("Page URL: %s" % wikipage.url)
print("Nr. of images on page: %d" % len(wikipage.images))
print(" - Main Image: %s" % wikipage.images[0])
```

```python
PAGE = "Brooklyn"
wikipage = wikipedia.page(PAGE)
img_url  = wikipage.images[0]
urllib.request.urlretrieve(img_url,
                           PROJECT_DIR + "www/img/" + PAGE + ".jpg")
print("File saved at:" + PROJECT_DIR + "www/img/" + PAGE + ".jpg")
```

```python
for city in data["City_Search_Bar"]:
    print(city)
    PAGE = city
    try:
        wikipage = wikipedia.page(PAGE)
        img_url  = wikipage.images[0]
        print("Page Title: %s" % wikipage.title)
        print("Page URL: %s" % wikipage.url)
        print("Nr. of images on page: %d" % len(wikipage.images))
        print(" - Main Image: %s" % img_url)
        urllib.request.urlretrieve(img_url, PROJECT_DIR + "www/img/" + PAGE + ".jpg")
        print("File saved at:" + PROJECT_DIR + "www/img/" + PAGE + ".jpg")
    except:
        pass
        print("")
```
